## Merge combine1_file + Movies title

```python
import os
import pandas as pd

# Define file path for combined_data_1.txt
file_path1 = os.path.join(path, "combined_data_1.txt")

# Initialize variables
movie_id = None
ratings_list = []

# Read the file and process data
with open(file_path1, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if line.endswith(":"):  # Movie ID line (e.g., "1:")
            movie_id = int(line[:-1])
        else:  # UserID, Rating, Date
            user_id, rating, date = line.split(",")
            ratings_list.append([int(user_id), movie_id, int(rating), date])

# Convert to DataFrame
df_ratings = pd.DataFrame(ratings_list, columns=["UserID", "MovieID", "Rating", "Date"])

# Save as CSV file
df_ratings.to_csv("/content/drive/MyDrive/Matrix/combined_data_1.csv", index=False)
print("Combined rating data saved as 'combined_data_1.csv'")
```

```python
# Load movie titles dataset
file_path2 = os.path.join(path, "movie_titles.csv")

df_movies = pd.read_csv(
    file_path2,
    encoding="ISO-8859-1",
    header=None,
    names=["MovieID", "Year", "Title"],
    quoting=3,  # Handle movie titles containing commas
    on_bad_lines="skip"
)

# Convert Year to integer type
df_movies["Year"] = df_movies["Year"].astype("Int64")

# Set MovieID as index
df_movies.set_index("MovieID", inplace=True)

# Save as CSV file
df_movies.to_csv("/content/drive/MyDrive/Matrix/movies_data.csv")

print("Movie information saved as 'movies_data.csv'")
```

```python
# Load saved rating and movie data
df_ratings = pd.read_csv("/content/drive/MyDrive/Matrix/combined_data_1.csv")
df_movies = pd.read_csv("/content/drive/MyDrive/Matrix/movies_data.csv", index_col="MovieID")

# Merge data
df_ratings = df_ratings.join(df_movies, on="MovieID")

# Save the final merged dataset
df_ratings.to_csv("/content/drive/MyDrive/Matrix/final_ratings.csv", index=False)

print(" Final merged data saved as 'final_ratings.csv'")

# Load final merged dataset
df_ratings = pd.read_csv("/content/drive/MyDrive/Matrix/final_ratings.csv")
df_ratings.head()
```
